## Question 1

Introduction:Special thanks to: https://github.com/justmarkham for sharing the dataset andmaterials.Occupations  
Step 1. Import the necessary libraries  
Step 2. Import the dataset from this address.  
Step 3. Assign it to a variable called users  
Step 4. Discover what is the mean age per occupation  
Step 5. Discover the Male ratio per occupation and sort it from the most to the least  
Step 6. For each occupation, calculate the minimum and maximum ages  
Step 7. For each combination of occupation and sex, calculate the mean age  
Step 8. For each occupation present the percentage of women and men

```python
#Step 1 :

import pandas as pd

#Step 2
url_lib = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

#Step 3
users = pd.read_csv(url_lib, sep='|')

#Step 4
mean_age_per_occup = users.groupby('occupation')['age'].mean()

#Step 5
male_ratio_per_occup = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').mean()).sort_values(ascending=False)

#Step 6
min_max_ages_per_occup = users.groupby('occupation')['age'].agg(['min', 'max'])

#Step 7
mean_age_per_combi = users.groupby(['occupation', 'gender'])['age'].mean()

#Step 8
total_count_per_occup = users.groupby('occupation')['gender'].count()
percentage_gender_per_occup = users.groupby(['occupation', 'gender'])['gender'].count() / total_count_per_occup * 100


print("Step 4 - Mean age per occupation ")
print(mean_age_per_occup)
print("\n")

print("Step 5 - Male ratio per occupation from the most to the least")
print(male_ratio_per_occup)
print("\n")

print("Step 6 - Minimum and maximum ages for each occupation")
print(min_max_ages_per_occup)
print("\n")

print("Step 7 - Mean age for each combination of occupation and sex ")
print(mean_age_per_combi)
print("\n")

print("Step 8 - percentage of women and man for each occupation")
print(percentage_gender_per_occup)
print("\n")

```

```python

dataset_path = r"C:\Users\mital\Desktop\BDAT DATA PROGRAMMING,ZIP FILE\Data set 3\wind.txt"
data = pd.read_table(dataset_path, sep="\s+")

data.index = pd.to_datetime(data.index)

def fix_year(x):
    return x.replace(year=1961)
data.index = data.index.map(fix_year)
  
data.index.name = None

nulls_per_location = data.isna().sum()
print("Number of Missing Values per Location:")
print(nulls_per_location)

total_values = data.count().sum()
print("\nTotal Number of Non-Missing Values:", total_values)

mean_all = data.mean().mean()
print("\nMean Windspeed Across All Locations and Times:", mean_all)

loc_stats = data.describe()
print("\nLocation Statistics:")
print(loc_stats)

day_stats = data.T.describe()
print("\n\n\nDay Statistics:")
print(day_stats)

jan_means = data[data.index.month==1].mean()
print("\n\n\nMean Windspeed in January for Each Location:")
print(jan_means)

yearly = data.resample('A').mean() 
print("\n\n\nYearly Downsampled Data:")
print(yearly)

monthly = data.resample('M').mean()
print("\n\n\nMonthly Downsampled Data:")
print(monthly)

weekly = data.resample('W-Mon').mean()
print("\n\n\nWeekly Downsampled Data:")
print(weekly)

weekly_crossloc = weekly.T.describe()
print("\n\n\nMin, Max, Mean, and Std across Locations for Each Week (First 52 Weeks):")
print(weekly_crossloc)
```

# Question 6 

Create a line plot showing the number of marriages and divorces per capita in theU.S. between 1867 and 2014. Label both lines and show the legend.Don't forget to label your axes

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
file_path =  r"C:\Users\mital\Desktop\BDAT DATA PROGRAMMING,ZIP FILE\Data set 3\us-marriages-divorces-1867-2014.csv"


df = pd.read_csv(file_path)

# Calculate marriages and divorces per 1000 people
df['Marriages_per_1000'] = df['Marriages'] / df['Population']
df['Divorces_per_1000'] = df['Divorces'] / df['Population']

# Plotting
plt.figure(figsize=(12, 8))

plt.plot(df['Year'], df['Marriages_per_1000'], label='Marriages per 1000', marker='o', linestyle='-', color='blue')
plt.plot(df['Year'], df['Divorces_per_1000'], label='Divorces per 1000', marker='o', linestyle='-', color='red')

plt.xlabel('Year')
plt.ylabel('Per 1000 People')
plt.title('Marriages and Divorces per 1000 People in the U.S. (1867-2014)')
plt.legend()

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Annotate the highest points
max_marriages = df.loc[df['Marriages_per_1000'].idxmax()]
max_divorces = df.loc[df['Divorces_per_1000'].idxmax()]

plt.annotate(f'Peak Marriages ({max_marriages["Year"]})', 
             xy=(max_marriages['Year'], max_marriages['Marriages_per_1000']),
             xytext=(max_marriages['Year'] + 10, max_marriages['Marriages_per_1000'] - 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.annotate(f'Peak Divorces ({max_divorces["Year"]})', 
             xy=(max_divorces['Year'], max_divorces['Divorces_per_1000']),
             xytext=(max_divorces['Year'] - 20, max_divorces['Divorces_per_1000'] + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```

# Question 7 

Create a vertical bar chart comparing the number of marriages and divorces percapita in the U.S. between 1900, 1950, and 2000.Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
dataset_path = r"C:\Users\mital\Desktop\BDAT DATA PROGRAMMING,ZIP FILE\Data set 3\us-marriages-divorces-1867-2014.csv"
df = pd.read_csv(dataset_path)

# Filter data for specific years
selected_years = [1900, 1950, 2000]
selected_data = df[df['Year'].isin(selected_years)].copy()

# Calculate per 1000 values
selected_data['Marriages_per_1000'] = selected_data['Marriages'] / selected_data['Population']
selected_data['Divorces_per_1000'] = selected_data['Divorces'] / selected_data['Population']

# Create the vertical bar chart
plt.figure(figsize=(10, 6))

x_values = selected_data['Year']
marriages_per_1000_values = selected_data['Marriages_per_1000']
divorces_per_1000_values = selected_data['Divorces_per_1000']

bar_width = 0.35
plt.bar(x_values - bar_width/2, marriages_per_1000_values, width=bar_width, label='Marriages per 1000')
plt.bar(x_values + bar_width/2, divorces_per_1000_values, width=bar_width, label='Divorces per 1000')

plt.xlabel('Year')
plt.ylabel('Per 1000')
plt.title('Comparison of Marriages and Divorces per 1000 in the U.S. (1900, 1950, 2000)')
plt.xticks(selected_years)
plt.legend()

plt.tight_layout()
plt.show()
```

# Question 8

# Question 9 

Create a pie chart showing the fraction of all Roman Emperors that wereassassinated.Make sure that the pie chart is an even circle, labels the categories, and shows thepercentage breakdown of the categories.

```python
# Step 1
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Read the dataset
emperor_dataset_path = r"C:\Users\mital\Desktop\BDAT DATA PROGRAMMING,ZIP FILE\Data set 3\roman-emperor-reigns.csv"
emperor_data = pd.read_csv(emperor_dataset_path)

# Step 3: Calculate the number of emperors who were assassinated
assassination_count = emperor_data[emperor_data['Cause_of_Death'] == 'Assassinated'].shape[0]

# Calculate the number of emperors who were not assassinated
non_assassination_count = emperor_data.shape[0] - assassination_count

# Step 4: Create the pie chart
categories = ['Assassinated', 'Not Assassinated']
counts = [assassination_count, non_assassination_count]
colors = ['limegreen', 'teal']
explode = (0.1, 0)

plt.figure(figsize=(8, 8))
plt.pie(counts, labels=categories, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)
plt.title('Fraction of Roman Emperors Assassinated')
plt.axis('equal') 

# Step 5: Display the plot
plt.show()
```
